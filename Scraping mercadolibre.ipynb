```python
def traer_infoCelular(url):
    datos = {}
    rs = requests.get(url)
    if rs.status_code == 200:
        bs = BeautifulSoup(rs.text)
        titulo = bs.find('div',attrs={'class':'ui-pdp-container__col col-2 mr-32'}).find('h1', attrs={'class':'ui-pdp-title'})
        if titulo: datos["titulos"] = titulo.text
        else: datos["titulos"] = None
        
        datos["5"] = bs.find('div',attrs={'class':'ui-review-view__rating__extended'}).find_all('div', attrs={'class':'ui-review-view-rating__level__column ui-review-view-rating__level__value-container'})[0].span.text
        datos["4"] = bs.find('div',attrs={'class':'ui-review-view__rating__extended'}).find_all('div', attrs={'class':'ui-review-view-rating__level__column ui-review-view-rating__level__value-container'})[1].span.text
        datos["3"]= bs.find('div',attrs={'class':'ui-review-view__rating__extended'}).find_all('div', attrs={'class':'ui-review-view-rating__level__column ui-review-view-rating__level__value-container'})[2].span.text
        datos["2"]= bs.find('div',attrs={'class':'ui-review-view__rating__extended'}).find_all('div', attrs={'class':'ui-review-view-rating__level__column ui-review-view-rating__level__value-container'})[3].span.text
        datos["1"]= bs.find('div',attrs={'class':'ui-review-view__rating__extended'}).find_all('div', attrs={'class':'ui-review-view-rating__level__column ui-review-view-rating__level__value-container'})[4].span.text
        
        
        
    return datos    
    
```

```python
traer_infoCelular(Lista_Links[0].get('href'))
```

```python
def traer_datos(links):
    info = []
    for e,i in enumerate(links):
        try:
            print(traer_infoCelular(i.get('href')))
            print("")
        except :
            print('no sirvio')
        print(f'ejecucion Num{e+1} de {len(links)} ')
    info

```

```python
traer_datos(Lista_Links)
```
